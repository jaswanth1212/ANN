```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
x = np.array([[2, 8], [3, 7], [4, 6], [5,5], [6,4], [1,10],[7,3], [8,2], [2,9], [9,1]])
y = np.array([[0], [0], [1], [1],[1], [0], [1],[1],[0],[1]])
model = Sequential([
    Dense(10, input_dim=2, activation='relu'),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x,y,epochs=100,verbose=0)
for i in x:
  prediction = model.predict(np.array([i]))
  print(f"input:{i} - prediction: {int(round(prediction.flatten()[0]))}")
```

```python
loss, train_acc = model.evaluate(x, y)
print(f"\nAccuracy: {train_acc:.4f}")
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
multi_layer=MLPClassifier(hidden_layer_sizes=(8,8,8),max_iter=100,random_state=25)
```

```python
multi_layer.fit(x_train,y_train)
```

```python
predict=multi_layer.predict(x_test)
```

```python
print(predict)
```

```python
accuracy=multi_layer.score(x_test,y_test)
print(accuracy)
```

```python
print(confusion_matrix(y_test,predict))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def linear(x):
  return x
def sigmoid(x):
  return 1/(1+np.exp(-x))
def leaky_relu(x):
  return np.maximum(0.1*x,x)
def relu(x):
  return np.maximum(0,x)
def hyperbolic_tangent(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def bipolar_sigmoid(x):
  return (1-np.exp(-x))/(1+np.exp(-x))
def softplus(x):
  return np.log(1+np.exp(x))


x = np.linspace(-10, 10, 100)
plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,linear(x))
plt.title('Linear')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,sigmoid(x))
plt.title('sigmoid')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,leaky_relu(x))
plt.title('leaky relu')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,relu(x))
plt.title('relu')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,hyperbolic_tangent(x))
plt.title('hyperbolic_tangent')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,bipolar_sigmoid(x))
plt.title('bipolar-sigmoid')
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(4,2,1)
plt.plot(x,softplus(x))
plt.title('softplus')
plt.show()





```
